```python
class Perceptron():   
    def __init__(self, rychlost_ucenia=0.01, pocet_iteracii=10): 
        # toto je konštruktor triedy, pri vytvarani inštancie triedy sa inicializuje
        #poceč iterácií a rýchlosť učenia
        self.eta = rychlost_ucenia 
        self.n_iter = pocet_iteracii      
    
    def fit(self, X,y): 
        self.vahy = np.zeros(1+ X.shape[1]) 
        print(self.vahy)
        print()
        #do premnennej vahy sa uloží jednorozmerne pole s troma nulami ([0. 0. 0.]),
        self.errors_=[]
        for i in range(self.n_iter): 
            errors = 0 
            for matica,vystup in zip(X,y): 
                print(self.vahy)
                #napríkld v prvom cykle: matica == [4.9 1.4] a vystup == -1
                update = self.eta * (vystup-self.predikcia(matica)) #určíme rozdiel vo výsledkoch
                self.vahy[1:] += update*matica # upravíme druhú a tretiu váhu 
                self.vahy[0]+= update #upravíme bias
                #napríklad v prvom cykle sú váhy nasledovné: 
                #pred vykonaním operácie váhy[1:] == [0. 0.], vahy[0] == 0.0
                #po vykonaní operácie váhy[1:] == [-0.098 -0.028], vahy[0] == -0.02
                errors += int(update !=0.0) 
            self.errors_.append(errors) # každú chybu pridáme do pola        
        return self  

    def net_input(self, X):
        #vynásobíme váhy som vstupom, sčítame a potom ešte priočítame bias 
        return np.dot(X, self.vahy[1:])+ self.vahy[0]


    def predikcia(self, X): 
        # táto funkcia volá metódu net_input a ak dostane naspať číslo
        # kt je vačšie alebo rovné 0, tak predikcia vráti 1 inak -1
        return np.where(self.net_input(X) >= 0.0, 1, -1) 
```
